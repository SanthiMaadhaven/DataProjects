```python
import matplotlib
import matplotlib.pyplot as plt
import numpy as np
import pandas as pd
```

### Download dataset


```python
weather_data=pd.read_csv('weather-1.csv')
weather_data.head()
```

#### Trim the dataset


```python
weather_data=weather_data[['Date','TempAvgF','WindAvgMPH']].head(30)
weather_data
```

### Plotting two lines with 2 separate Y axes with the same X axis


```python
#Configure the figure
fig,ax_tempF=plt.subplots()
fig.set_figwidth(12)
fig.set_figheight(6)

#Configure the X axis
ax_tempF.set_xlabel('DAte')
ax_tempF.tick_params(axis='x',bottom=False,labelbottom=False)



#Configure the first Y axis
ax_tempF.set_ylabel('Temp(F)',size='x-large')
ax_tempF.tick_params(axis='y',labelsize='large')

#first plot date vs TEMP
ax_tempF.plot(weather_data['Date'],weather_data['TempAvgF'],color='r')


# twinx() for dual axis
ax_dual=ax_tempF.twinx()


#Configure the second Y axis
ax_dual.set_ylabel('avg wind speed',size='x-large')
ax_dual.tick_params(axis='y',labelsize='large')

#second plot date vs windspeed
ax_dual.plot(weather_data['Date'],weather_data['WindAvgMPH'],color='g')


```

### Twin axis for the same data


#### Define function to convert from fahrenheit to celsius

```python
def fahrenheit2celsius(f):
    return (f-32)*5/9
    
```

```python
#Configure the figure
#Configure the figure
fig,ax_tempF=plt.subplots()
fig.set_figwidth(12)
fig.set_figheight(6)

#Configure the X axis
ax_tempF.set_xlabel('DAte')
ax_tempF.tick_params(axis='x',bottom=False,labelbottom=False)



#Configure the first Y axis
#Configure the first Y axis
ax_tempF.set_ylabel('Temp(F)',size='x-large')
ax_tempF.tick_params(axis='y',labelsize='large')



#first plot date vs temp - F
#first plot date vs TEMP
ax_tempF.plot(weather_data['Date'],weather_data['TempAvgF'],color='r')


# twinx() for dual axis
ax_tempc=ax_tempF.twinx()


#Configure the second Y axis
#Configure the second Y axis
ax_tempc.set_ylabel('Temp(c)',size='x-large')
ax_tempc.tick_params(axis='y',labelsize='large')


#second axis configure: temp - c
ymin,ymax=ax_tempF.get_ylim()
ax_tempc.set_ylim(fahrenheit2celsius(ymin),fahrenheit2celsius(ymax))




```
