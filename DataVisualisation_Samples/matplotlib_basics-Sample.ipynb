# add_axes


```python
# places a figure in the canvas that is exactly as large as the canvas itself.
#add_axes(rect), where rect is a list [x0, y0, width, height]

fig=plt.figure()
ax=fig.add_axes([0,0,1,1])
```

```python
#add two vertical axes and plots
fig=plt.figure()
ax1=fig.add_axes([0,0.6,1,0.4])
ax2=fig.add_axes([0,0,0.8,0.4])

x=np.linspace(start=0,stop=10,num=50)

ax1.plot(x,np.sin(x))
ax2.plot(x,np.cos(x))


```

# subplot2grid


```python

ax1=plt.subplot2grid((2,3),(0,0))
ax1.plot(x,np.sin(x))

ax2=plt.subplot2grid((2,3),(0,1))
ax2.plot(x,np.sin(x))

ax3=plt.subplot2grid((2,3),(0,2),rowspan=2)
ax3.plot(x,np.sin(x))

ax4=plt.subplot2grid((2,3),(1,0),colspan=2)
ax4.plot(x,np.sin(x))
ax4.set_title("4th plot")



```
