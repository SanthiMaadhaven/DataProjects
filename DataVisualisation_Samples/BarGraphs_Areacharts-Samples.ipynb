# Bar chart

## Create a simple bar chart

```python
import matplotlib.pyplot as plt
import numpy as np
x=np.arange(4)
y=[12,34,31,20]
```

```python
plt.bar(x,y,width=0.8)
plt.xticks(np.arange(4),('a','b','c','d'))
```

__Plotting multiple sets of data__<br>
__Hiding Y axis__


```python
x1=[1,3,5,7]
x2=[2,4,6,8]
y1=[7,7,7,7]
y2=[17,18,29,40]

b1=plt.bar(x1,y1,label='first bar')

b2=plt.bar(x2,y2,label='second bar')

plt.legend()

for bar1 in b1:
    yval=bar1.get_height()
    plt.text(bar1.get_x()+(bar1.get_width()/2),yval,int(yval),va='bottom') 

```

# Horizontal bar graph

```python
X=[28,33,43,45,57]
Y=["A","B","C","D","E"]

plt.barh(Y,X)
```

Sort the bars<br>
Reverse the order

```python
b=zip(X,Y)
a=sorted(zip(X,Y))



y,x=zip(*sorted(zip(Y,X)))
print(list(b),a,y,x)
plt.barh(y,x)

ax=plt.subplot()
ax.invert_yaxis()
```

# Grouped Bar chart

```python
year1=(20,35,30,35,27,27,33)
year2=(25,39,45,25,21,29,36)


x=np.arange(7)
width=0.45

fig,ax=plt.subplots()
r1=ax.bar(x-width/2,year1,width,color='b', label='year1',alpha=0.8)
r2=ax.bar(x+width/2,year2,width,color='g', label='year2', alpha=0.2)

plt.legend()

```

# Stacked bar chart


```python
bottom_array=(20,35,30,35)
top_array=(25,32,34,15)

rect1=plt.bar(np.arange(4),bottom_array,width=0.5,color='r')

rect2=plt.bar(np.arange(4),top_array,bottom=bottom_array,width=0.5,color='g')
```

```python
y=['a','b','c','d']
plt.barh(y,bottom_array)
plt.barh(y,top_array,left=bottom_array)
```
