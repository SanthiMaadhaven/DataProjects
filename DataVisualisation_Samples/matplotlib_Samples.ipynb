```python
import matplotlib
import matplotlib.pyplot as plt
import numpy as np
```

# Decoration with Plot Styles and Types

```python
array=np.random.randn(20)
plt.plot(array,color="r",linestyle="--",linewidth=5)


```

# adding text to the plot

```python
x=[2,4,6,8,10]
y=[4,8,12,16,20]

plt.plot(x,y)
plt.text(2,4,"this is the text at (2,4)")
plt.text(6,8,"this is the text at (4,8)")



```

# adding annotations to the plot

```python
x=[2,4,6,8,10]
y=[4,8,12,16,20]

plt.plot(x,y,'bo-')
plt.grid()

for x,y in zip(x,y):
     label="{:.2f}".format(y)
     plt.annotate (label,
                   (x,y),
                   textcoords="offset points",
                   xytext=(0,10),
                   ha="center" 
                   )


```

```python
xs=[2,4,6,8,10]
ys=[4,8,12,16,20]

plt.scatter(xs,ys)


```

# Scatterplot

# Logarithmic Axes

```python
fig=plt.figure()


plot1=fig.add_subplot(2,1,1)
plot1.plot([3,4,5],[10,20,30])
plot1.set_ylabel("normal scale")

plot2=fig.add_subplot(2,1,2)
plot2.plot([3,4,5],[10,20,30])
plot2.set_yscale('log')
plot2.set_ylabel("log scale")



```

```python
#Share axis
fig=plt.figure()


plot1=fig.add_subplot(2,2,1)
plot1.plot([3,4,5],[10,20,30])

plot2=fig.add_subplot(2,2,3,sharex=plot1)
plot2.plot([3,4,5],[10,20,30])

plt.subplots_adjust(wspace=0,hspace=0)





```

```python
# Remove top and right frame/lighten the x axis
ax=plt.axes()

plt.plot([3,4,5],[10,20,30])

# Hide the right and top spines
ax.spines['right'].set_visible(False)
ax.spines['top'].set_visible(False)
ax.spines['bottom'].set_color('grey')





```
