# Figure-level  functions

Figure-level functions interface with matplotlib through a seaborn object, usually a FacetGrid, that manages the figure. Each module has a single figure-level function, which offers a unitary interface to its various axes-level functions. 



Visualizing distributions of data
displot

Figure-level interface for drawing distribution plots onto a FacetGrid.

histplot

Plot univariate or bivariate histograms to show distributions of datasets.

kdeplot

Plot univariate or bivariate distributions using kernel density estimation.

ecdfplot

Plot empirical cumulative distribution functions.

rugplot

Plot marginal distributions by drawing ticks along the x and y axes.

distplot

DEPRECATED: Flexibly plot a univariate distribution of observations.

### displot()

It is figure-level function for the distributions module (default - histplot()). 



```python
#stacked histogram using distplot
sns.displot(data=penguins, y='flipper_length_mm',hue='sex')
```

```python
#separate col and dodged
sns.displot(data=penguins, y='flipper_length_mm',col='sex')
```

### Multiple subplots using distpot
    facet them by plotting each distribution across the columns of the figure:

```python
sns.displot(data=penguins, y='flipper_length_mm',col='sex')
```

# A bivariate distribution plot


```python
#bivariate distplot
sns.displot(data=penguins, x='bill_length_mm', y='bill_depth_mm')
```

```python
#colour bar
sns.displot(data=penguins, x='bill_length_mm', y='bill_depth_mm',cbar=True)
```

### Normalized histogram statistics


```python
#normalize the counts using the stat parameter.
#Density normalization scales the bars so that their areas sum to 1
sns.displot(data=penguins, x='flipper_length_mm',hue='species',stat='density')
```

```python
#normalize the bars to that their heights sum to 1
sns.displot(data=penguins, x='flipper_length_mm',hue='species',stat='probability')
```

# kde using distplot (kind parameter)

```python
#kde using distplot (kind parameter)
sns.displot(data=penguins, x='flipper_length_mm',hue='species',kind='kde')
```

```python
#add hue,col

sns.displot(data=penguins, x='flipper_length_mm',hue='species',kind='kde',col='sex')
```

```python
#Stacked KDE
sns.displot(data=penguins, x='flipper_length_mm',hue='species',kind='kde',multiple='stack')
```

```python
#histogram+KDE
sns.displot(data=penguins, x='flipper_length_mm',hue='species',kde=True)
```

```python
#bivariate KDE
sns.displot(data=penguins, x='bill_length_mm', y='bill_depth_mm',kind='kde')
```

```python
#using axis level function kdeplot
sns.kdeplot(data=penguins, x='flipper_length_mm',hue='species',multiple='stack')
```

# Plotting joint and marginal distributions

```python
#jointplot
sns.jointplot(data=penguins, x='bill_length_mm', y='bill_depth_mm')
```

```python
#jointplot KDE
sns.jointplot(data=penguins, x='bill_length_mm', y='bill_depth_mm',kind='kde',hue='sex')
```

```python
#jointgrid (Hist and box)
g=sns.JointGrid(data=penguins, x='bill_length_mm', y='bill_depth_mm')
g.plot_joint(sns.histplot)
g.plot_marginals(sns.boxplot)
```

## Plot pairwise relationships in a dataset -pairplot()

multiple pairwise bivariate distributions in a dataset

```python
#pairplot()
sns.pairplot(penguins)
```

```python
#KDE pairplots
sns.pairplot(penguins,kind='kde')
```

```python
#hist pariplot
```

```python
#Assign hue - defalt KDE plot for marginal distribution
sns.pairplot(penguins,hue='species')
```

```python
# pairplot - histogram for marginal distribution 
sns.pairplot(penguins,hue='species',diag_kind='hist')
```

```python
#pairplot for selected variables
sns.pairplot(penguins,
             x_vars=['bill_length_mm','bill_depth_mm'],
             y_vars=['bill_length_mm','bill_depth_mm','flipper_length_mm']
            
            )
```

### pairgrid

```python
#blank grid 
sns.PairGrid(penguins)
```

```python
#pairgrid-scatterplot
g=sns.PairGrid(penguins)
g.map(sns.scatterplot)
```

```python
#map_diag - histpgram, map_offdiag-scatterplot
g=sns.PairGrid(penguins)
g.map_diag(sns.histplot)
g.map_offdiag(sns.scatterplot)
```

```python
#upper, lower and diagonal elements
#map_diag - histpgram, map_offdiag-scatterplot
g=sns.PairGrid(penguins)
g.map_diag(sns.histplot,kde=True)
g.map_upper(sns.histplot)
g.map_lower(sns.kdeplot)
```
