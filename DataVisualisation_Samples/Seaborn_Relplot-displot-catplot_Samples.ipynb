```python
import pandas as pd 
import numpy as np
import matplotlib as mpl
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
sns.get_dataset_names()
```

```python
iris = sns.load_dataset("iris")
```

```python
iris.shape
```

```python
iris.head()
```

```python
iris.species.value_counts()
```

```python
iris.columns
```

```python
sns.displot(iris,x="sepal_length")
```

```python
sns.displot(iris,x="sepal_length",binwidth=0.2)
plt.show()
```

```python
sns.displot(iris,x="sepal_length",binwidth=0.2,hue="species")
plt.show()
```

```python
sns.displot(iris,x="sepal_length",binwidth=0.2,hue="species",col="species")
plt.show()
```

```python
sns.displot(iris,x="sepal_length",binwidth=0.2,hue="species",multiple="dodge")
plt.show()
```

```python
sns.displot(iris,x="sepal_length",hue="species",kind="kde")
plt.show()
```

```python
sns.catplot(data=iris,x="species",y="sepal_length",kind="box")
plt.show()
```

```python
iris.columns
```

```python
sns.relplot(x="sepal_length",y="petal_length",data=iris,hue="species")
plt.show()
```

```python
sns.relplot(x="sepal_length",y="petal_length",data=iris,hue="species",col="species")
plt.show()
```

```python
sns.jointplot(x="sepal_length",y="petal_length",data=iris,hue="species")
plt.show()
```

```python
sns.pairplot(data=iris,hue="species")
plt.show()
```

```python
fig,ax = plt.subplots(2,2,figsize=(18,8))
sns.boxplot(data=iris,x="species",y="sepal_length",ax=ax[0][0])
sns.violinplot(data=iris,x="species",y="sepal_length",ax=ax[0][1])
sns.kdeplot(data=iris,x="sepal_length",hue="species",ax=ax[1][0])
sns.scatterplot(x="sepal_length",y="petal_length",data=iris,hue="species",ax=ax[1][1])
plt.show()
```
